```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
URL = "https://www.theglobaleconomy.com/rankings/wb_political_stability/"
r = requests.get(URL)
```

```python
soup = BeautifulSoup(r.content)
table = soup.find('table')

df = pd.read_html(str(table))[0]

# Set the maximum number of rows to be displayed to the number of rows in the DataFrame
pd.set_option('display.max_rows', df.shape[0])

df
```

```python
# Drop the fourth column
df = df.drop(df.columns[3], axis=1)

df
```

```python
# Create a boolean mask that selects the rows to be dropped
mask = (df['Countries'] != 'USA') & (df['Global rank'] % 2 == 0)
```

```python
# Drop the rows that match the boolean mask
df1 = df.drop(df[mask].index)

df1
```

```python
# Create a boolean mask that selects the rows to be dropped-this is to reduce the amount of data that will be visualised
mask = df['Countries'].str.startswith(('A', 'M', 'S', 'Z', 'B', 'I', 'F', 'T'))

# Drop the rows that match the boolean mask
df2 = df1.drop(df1[mask].index)

df2
```

```python
# Create a boolean mask that selects the rows to be dropped
mask = df['Countries'].str.contains(' ', '.')

# Drop the rows that match the boolean mask
df3 = df2.drop(df2[mask].index)

df3
```

```python
# Save the DataFrame as a CSV file
df3.to_csv('political_stability.csv', index=False)
```
