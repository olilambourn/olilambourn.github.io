```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

URL = "https://www.teachingabroaddirect.co.uk/blog/student-budgets-around-the-world"
r = requests.get(URL)
  
soup = BeautifulSoup(r.content)
table = soup.find('table')

df = pd.read_html(str(table))[0]
df.head()
```

```python
# Select rows where the 'Country' column is 'China'
china_df = df.loc[df['Country'] == 'China']

```

```python
# Use the melt function to reformat the DataFrame
melted_df = pd.melt(china_df, id_vars=['Country'], value_vars=['2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018'], var_name='Year', value_name='Value')
```
